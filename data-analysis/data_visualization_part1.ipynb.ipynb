```python
#Histogram
np.random.seed(42)
scores = np.random.normal(60, 10, 100)
plt.figure(figsize=(10,6))
plt.hist(scores, bins = 10, color = 'skyblue', alpha = 0.7, edgecolor = 'black')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of scores')
plt.axvline(scores.mean(), color = 'red', linestyle = '--', 
           label = f'Mean: {scores.mean():.1f}')
plt.axvline(scores.max(), color = 'green', linestyle = '-', label = f'Max: {scores.max():.1f}')
plt.axvline(scores.min(), color = 'purple', linestyle = '-', label = f'Min: {scores.min():.1f}')
plt.legend()
plt.show()
```

```python
import pandas as pd
import folium
```

```python
data = {
    'city': ['New York', 'London', 'Tokyo', 'Sydney', 'Cairo'],
    'country': ['USA', 'UK', 'Japan', 'Australia', 'Egypt'],
    'lat': [40.7128, 51.5074, 35.6895, -33.8688, 30.0444],      # Latitudes
    'lon': [-74.0060, -0.1278, 139.6917, 151.2093, 31.2357]     # Longitudes
}
df = pd.DataFrame(data)
df.head()
```

```python
map = folium.Map(location = [20,0], zoom_start = 2)
map
```

```python
for index, row in df.iterrows():
    folium.Marker(
        location = [row['lat'], row['lon']],
        popup = f"{row['city']}, {row['country']}",
        icon = folium.Icon(color = 'red', icon = 'info-sign')
    ).add_to(map)
map
```
