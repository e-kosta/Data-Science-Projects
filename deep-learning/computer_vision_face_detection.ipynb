```python
img = cv2.imread('image.jpg')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB )
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(gray)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.imshow(hsv)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
overlay = np.zeros_like(img)
overlay[:,:]= [50,0,100]
```

```python

blended = cv2.addWeighted(img, 0.7, overlay, 0.3, 0)
```

```python
bright = cv2.add(img, np.array([50.0]))
```

```python
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
plt.figure(figsize = (12,5))
plt.imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

Drawing on Images

```python
canvas = img.copy()
cv2.rectangle(canvas, (50,50), (200,200), (0,255,0), 3)
cv2.circle(canvas, (300,300), 50, (255,0,0), -1)
cv2.line(canvas, (0,0), (500,500), (0,0,255), 3)
cv2.putText(canvas, 'Computer Vision', (50,400), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
plt.figure(figsize = (10,8))
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# LOAD HAAR CASCADE FOR FACE DETECTION 
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades
                                    + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
```

```python
# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(canvas, (x, y), (x+w, y+h), (0, 255, 0), 3)
```

```python
plt.figure(figsize = (10,8))
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
