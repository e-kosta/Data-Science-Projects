```python
word1 = nlp("Artificial")
word2 = nlp("intelligence")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine similarity = A*B #dot product cosX 
                    /\A\*\B\
```

```python
similarity = cosine_similarity([word1.vector], [word2.vector])
print("Cosine similarity between Artificial and intelligence:", similarity[0][0])
```
