### **<font color='#306998'>TASK </font><font color='#ffd33b'>1.16</font>**

Write a function `get_sum_of_squared_naturals()` that takes one argument as an input (call it `max_value`) and returns the sum of all natural numbers such that $k \leq max\_value.$

Make sure that the input given by a user is a natural number. Set the default value to 5.

````python
def get_sum_of_squared_naturals(max_value=5):
    if not isinstance(max_value, int) or max_value<=0:
        raise ValueError("Input must be a natural number(positive integer).")
    
    sum_of_squares=sum(k**2 for k in range(1, max_value+1))
    
    return sum_of_squares

result=get_sum_of_squared_naturals(5)
print(result)
````

Write a function that accepts a positive natural number $p\in\mathbb{N}_1$ and two NumPy vectors as an input (both of the same shape). The function should return the Mikowski distance between those vectors, ie.:
$$d_p(X, Y) = \left(\sum_{i=1}^n |x_i - y_i|^p\right)^\frac{1}{p},$$

where $X=[x_1, x_2, ..., x_n]\in\mathbb{R}^n$ and $Y=[y_1, y_2, ..., y_n]\in\mathbb{R}^n.$

*Hint: you can use the following assertions `assert a.size == b.size, "Arrays must be of the same shape"`, and `assert isinstance(input_array_1, np.ndarray)`.*

````python
def minkowski_distance (X,Y,p):
    #X(numpy.ndarray): first vector
    #Y: second vector
    #P(int): Order of the Minkowski distance 
    assert isinstance(X, np.ndarray), "X must be a NumPy array"
    assert isinstance(Y, np.ndarray), "Y must be a NumPy array"
    assert X.size==Y.size, "Vectors must have the same shape"
    assert p>=1, "p must be a positive integer"
    #calculate the minkowski distance 
    distance=np.sum(np.abs(X-Y)**p)**(1/p)
    return distance

X=np.array([1,2,3])
Y=np.array([4,5,6])
p=4

result= minkowski_distance(X,Y,p)
print("Minkowski distance:", result)
````

### **<font color='#306998'>TASK </font><font color='#ffd33b'>1.26</font>**

For the following vectors `a=[3, -2]` and `b=[-2, -3]`, determine their scalar product and the angle between these vectors.

*Correct answer $\alpha = 90^\circ$.*

````python
a=np.array([3,-2])
b=np.array([-2,-3])

dot_product=np.dot(a,b)

magnitude_a=np.linalg.norm(a)
magnitude_b=np.linalg.norm(b)

angle_radians=np.arccos(dot_product/(magnitude_a*magnitude_b))

angle_degrees=np.degrees(angle_radians)

print("The angle between the two vectors is:", angle_degrees)
````

<span style="color:red;">
The binomial distribution models the number of successes in 
n independent trials, where each trial has a probability p of success. The distribution becomes skewed depending on the value of p:

````python
data = load_iris()
X = data.data
y = data.target
target_names = data.target_names

iris_df = pd.DataFrame(X, columns=data.feature_names)
iris_df['species'] = [target_names[i] for i in y]

setosa_sepal_length = iris_df[iris_df['species'] == 'setosa']['sepal length (cm)']
virginica_sepal_length = iris_df[iris_df['species'] == 'virginica']['sepal length (cm)']

plt.figure(figsize=(10, 5))
sns.histplot(setosa_sepal_length, bins=10, color='lightblue', label='Iris Setosa', kde=True)
sns.histplot(virginica_sepal_length, bins=10, color='lightcoral', label='Iris Virginica', kde=True)
plt.title('Distribution of Sepal Lengths for Iris Setosa and Virginica')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()

ks_stat, ks_p = ks_2samp(setosa_sepal_length, virginica_sepal_length)
print(f'Kolmogorov-Smirnov Test Statistic: {ks_stat}, p-value: {ks_p}')

alpha = 0.05
if ks_p > alpha:
    print("Fail to reject the null hypothesis: The samples are drawn from the same population.")
else:
    print("Reject the null hypothesis: The samples are drawn from different populations.")
````

<span style="color:red">A low p-value (typically below 0.05) indicates that you can reject the null hypothesis, suggesting a significant difference between the sample mean and the population mean.
